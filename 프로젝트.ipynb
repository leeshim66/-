```python
#################### 클러스터링 
from sklearn.decomposition import PCA

pca = PCA(n_components=5)

#fit( )과 transform( )  PCA 변환 데이터 반환
pca.fit(X_minmax)
data_pca = pca.transform(X_minmax)
print(data_pca.shape)
pca.explained_variance_ratio_.cumsum()
```

```python
pca.components_.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
coef = pd.Series(pca.components_[0], index=X.columns)
coef_sort = coef.sort_values(ascending=False)[:10]
plt.figure(figsize=(10,10))
sns.barplot(x=coef_sort.values, y=coef_sort.index)
plt.title("첫번째 PCA component의 변수별 가중치")
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
km = KMeans(n_clusters=2) # 실제 집단은 2가지(이식의료기관, 일반병원)이므로 n_clusters를 2로 지정
km.fit(pca_df.iloc[:,:-1].values) # 마지막 열은 실제 target value이므로 제외
pred = km.predict(pca_df.iloc[:,:-1].values) 

origin_zero = [] # 일반 병원인데 이식의료기관으로 분류된 병원 index
for i in range(len(y)):
    if y_all[i]==0 and pred[i]==1:
        origin_zero.append(i)

최종엑셀.iloc[origin_zero,[0,2]]
```
